## 1. 과일 가격 출력하기


**문제 설명:**

과일 이름이 담긴 리스트가 주어집니다. for문을 사용하여 각 과일을 순서대로 출력하되, "N번째: 과일이름" 형식으로 출력하세요. (N은 1부터 시작)

**입력 예시:**

```
입력 없음
```

**출력 예시:**
```
1번째: 사과
2번째: 바나나
3번째: 포도
4번째: 오렌지
```

````python
fruits = ["사과", "바나나", "포도", "오렌지"]

for i in range(len(fruits)) :
    print(f"{i+1}번째: {fruits[i]}")
````


## 5. 전화번호부 만들기


**문제 설명:**

3명의 이름과 전화번호를 입력받아 딕셔너리에 저장하세요. 그 후 사용자로부터 이름을 입력받아 해당 이름의 전화번호를 출력하세요. 이름이 없으면 "등록되지 않은 이름입니다."를 출력하세요.

**입력 예시:**

```
김철수
010-1234-5678
이영희
010-2345-6789
박민수
010-3456-7890

이영희
```

**출력 예시:**
```
이영희의 전화번호: 010-2345-6789
```

**입력 예시:**

```
김철수
010-1234-5678
이영희
010-2345-6789
박민수
010-3456-7890

홍길동
```

**출력 예시:**
```
등록되지 않은 이름입니다.
```

````python
# 전화번호부 입력
phone_book = {}
name=[]
numbers=[]
for i in range(3) :
    name.append(input('이름을 입력하세요'))
    numbers.append(input('사용자 번호를 입력하세요(-포함)'))

phone_book=dict(zip(name,numbers))

# 사용사 이름 검색
search = input('이름을 검색하세요')

if search in phone_book :
    print(f'{search}의 전화번호: {phone_book[search]}')
else :
    print('등록되지 않은 이름입니다.')
````


## 6. 재고 관리 시스템 (딕셔너리 + 반복문)


**문제 설명:**

상품명을 key, 재고 수량을 value로 하는 딕셔너리가 주어집니다.

- 전체 재고를 출력하세요.
- 재고가 10개 미만인 상품을 찾아 "부족" 표시와 함께 출력하세요.
- 전체 재고의 총 합계를 출력하세요.

**입력 예시:**

```
입력 없음
```

**출력 예시:**
```
=== 전체 재고 현황 ===
노트북: 15개
마우스: 8개
키보드: 12개
모니터: 5개
헤드셋: 20개

=== 재고 부족 상품 ===
마우스: 8개 (부족)
모니터: 5개 (부족)

총 재고: 60개
```

````python
# 1) 재고 딕셔너리 만들기
inventory = {
    "노트북": 15,
    "마우스": 8,
    "키보드": 12,
    "모니터": 5,
    "헤드셋": 20
}

# 2) 전체 재고 출력 (딕셔너리 + for 반복)
print("=== 전체 재고 현황 ===")
for product in inventory:                 # product는 key(상품명)
    qty = inventory[product]              # qty는 value(수량)
    print(f"{product}: {qty}개")

print()

# 3) 재고 부족(10개 미만) 상품만 출력
print("=== 재고 부족 상품 ===")
for product in inventory:
    qty = inventory[product]
    if qty < 10:
        print(f"{product}: {qty}개 (부족)")

print()

# 4) 총 재고 합계 (반복문으로 직접 더하기)
total = 0
for product in inventory:
    total = total + inventory[product]
print(f"총 재고: {total}개")
````


## 7. 짝수/홀수 분류

**문제 설명:**

1부터 20까지의 숫자 중에서 짝수만 출력하되, 10 이상인 짝수는 "큰 짝수: N"으로, 10 미만인 짝수는 "작은 짝수: N"으로 출력하세요.

**입력 예시:**

```
입력 없음
```

**출력 예시:**
```
작은 짝수: 2
작은 짝수: 4
작은 짝수: 6
작은 짝수: 8
큰 짝수: 10
큰 짝수: 12
큰 짝수: 14
큰 짝수: 16
큰 짝수: 18
큰 짝수: 20
```

````python
i = 0

while i <20 :
    if i >=10  and i%2==0 :
        print(f"큰 짝수: {i}")
    elif i >0 and i%2==0:
        print(f"작은 짝수: {i}")
    i+=1
````


## 8. 업다운 게임

**문제 설명:**

정답 숫자를 미리 정해두고(예: 42), 사용자가 숫자를 입력할 때마다 "Up" 또는 "Down" 힌트를 주세요. 정답을 맞추면 "정답입니다! N번 만에 맞추셨습니다."를 출력하고 종료하세요.

Hint: while문 + break

**입력 예시:**

```
정답: 42

50
30
40
42
```

**출력 예시:**
```
Down!
Up!
Up!
정답입니다! 4번 만에 맞추셨습니다.
```

````python
# 원하는 정답 숫자를 하나 정하세요!
answer = int(input('정답: '))

i=1

while True :
   
    num = int(input('정답을 입력해주세요'))
    
    if num == answer :
        print(f'정답입니다! {i}번 만에 맞추셨습니다.')
        break
    elif answer > num :
        print('Up!')
        i+=1
    elif answer < num :
        print('Down!')
        i+=1

````


## 9. ATM 시뮬레이션

**문제 설명:**

초기 잔액 100,000원으로 시작하는 ATM 프로그램을 while문 무한 루프와 break를 활용하여 만드세요. 각 메뉴에 해당하는 번호를 입력 받고:

1. 현재 잔액 조회
2. 입금: 입금액을 추가로 입력받습니다. (잔액에 추가)
3. 출금: 출금액을 추가로 입력받습니다. (잔액에서 차감) (잔액 부족 시 에러 메시지)
4. 종료

"4"를 입력하면 프로그램을 종료합니다.

**입력 예시:**

```
1
2
50000
1
3
30000
1
4
```

**출력 예시:**
```
=== ATM ===
1. 잔액 조회
2. 입금
3. 출금
4. 종료
선택: 1
현재 잔액: 100000원

선택: 2
입금액: 50000
입금 완료! 현재 잔액: 150000원

선택: 1
현재 잔액: 150000원

선택: 3
출금액: 30000
출금 완료! 현재 잔액: 120000원

선택: 1
현재 잔액: 120000원

선택: 4
ATM을 종료합니다.
```

````python
#1. 현재 잔액 조회
#2. 입금: 입금액을 추가로 입력받습니다. (잔액에 추가)
#3. 출금: 출금액을 추가로 입력받습니다. (잔액에서 차감) (잔액 부족 시 에러 메시지)
#4. 종료 "4" 를 입력하면 프로그램을 종료합니다.

money = 100000 

 
while True :
    choice = int(input("번호를 선택하세요 "))
    print("=== ATM ===")
    print("1. 잔액 조회")
    print("2. 입금")
    print("3. 출금")
    print("4. 종료")

    if choice ==1 :
        print(f"선택{choice}")
        print(f"현재잔액{money}")
    elif choice == 2:
        print(f"선택{choice}")

        input_money = int(input('입금액: '))
        print(f"입금액{input_money}")
        money+= input_money
        
        print(f"입금 완료! 현재 잔액: {money}원")

    elif choice == 3 :
        print(f"선택{choice}")

        out_money = int(input('출금액'))
        print(f"출금액{out_money}")
        money-= out_money
        
        print(f"출금 완료! 현재 잔액: {money}원")
    elif choice == 4:
        print(f"선택{choice}")
        print("ATM을 종료합니다.")
        break



````
